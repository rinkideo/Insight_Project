### Creating a DATABASE from the CSV File

```python
##CREATE A CONNECTION TO A DATABASE
birth_database = create_engine('sqlite:///birth_database.db')
```

```python
#BUILDING THE DATABASE BY CHUNKING
chunksize=10000
i=0
j=0


for df in pd.read_csv(file,chunksize=chunksize, iterator=True,encoding ='unicode_escape',engine='python'):
    df = df.rename(columns= {c: c.replace(' ', '') for c in df.columns})
    conn = sqlite3.connect('birth_database.db') 
    conn.text_factory = lambda x: unicode(x, 'utf-8', 'ignore')
    df.index += j
    df.to_sql('birth1_2017', birth_database, if_exists='append')
    j=df.index[-1]+1
    print('| index {}'.format(j))
```
